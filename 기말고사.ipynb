## vaccines 컬럼과 country 컬럼을 참고하여 백신별 접종 국가수

```python
new_df = df[['country','vaccines']]
new_df
```

```python
vaccines_li = df['vaccines'].tolist()
vaccines = []
for vaccine in vaccines_li:
    if vaccine not in vaccines:
        vaccines.append(vaccine)
vaccines
```

```python
vaccine_list = []
for vaccine in vaccines:
    vs = vaccine.split(', ')
    for v in vs:
        if v not in vaccine_list:
            vaccine_list.append(v)
vaccine_list
```

```python
new_df.shape[0]
```

```python
vaccine_dict = {}
for vaccine in vaccine_list:
    vaccine_dict[vaccine] = 0
vaccine_dict
```

```python
dict = {}
for vaccine in vaccine_list:
    for idx in range(new_df.shape[0]):
        if vaccine in df.loc[idx,'vaccines']:
            vaccine_dict[vaccine] += 1
vaccine_dict
```

```python
vaccine_types = list(vaccine_dict.keys())
```

```python
vaccine_country = list(vaccine_dict.values())
```

```python
vaccine_country
```

```python
result = pd.DataFrame(vaccine_types)
result
```

```python
result['접종 국가수'] = vaccine_country
result
```

```python
result.columns = ['백신 이름', '백신 국가수']
result
```
