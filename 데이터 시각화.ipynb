```python
import matplotlib.pyplot as plt
```

## 기본 그래프 그리기

```python
xdata = [1,2,3,4]
ydata = [10,5,20,50]

plt.plot(xdata, ydata)
plt.show()
```

```python
xdata = [1,2,3,4]
ydata = [10,25,37,50]

plt.bar(xdata, ydata)
plt.show()
```

```python
plt.barh(xdata, ydata)
plt.show()
```

```python
plt.hist([1,1,2,3,4,5,6,6,7,8,10])
plt.show()
```

```python
data = [10,20,30]
plt.pie(data)
plt.show()
```

```python
xdata = [1,2,3,4]
ydata = [10,40,20,30]
plt.scatter(xdata, ydata)
plt.show()
```

## 그래프 그리기 옵션

```python
data1 = [1,2,3,4]
data2 = [4,3,1,1]

plt.title('plot graph')
plt.plot(data1, 'r.', label="circle")  # 색상: r,g,b 모양: ., ^
plt.plot(data2, 'g^', label='triangle')
plt.legend()  # 범례 표시
plt.show()
```

```python
data = [10,30,20,45]
label = ['A', 'B', 'C', 'D']
color = ['hotpink', 'darkmagenta', 'pink', 'deeppink']

plt.pie(data, labels=label, autopct='%.1f%%', colors=color, explode=(0, 0, 0.1, 0))  # autopct: 비율, explode: 돌출
plt.legend()
plt.show()
```

```python
xdata = [1,2,3,4]
ydata = [10,40,20,30]
sizes = [10,50,100,200]

plt.scatter(xdata, ydata, s=sizes, c=range(4), cmap='jet')  # s: 크기, c: 색상 범위, cmap: 색상 맵
plt.colorbar()
plt.show()
```
